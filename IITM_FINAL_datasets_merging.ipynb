```python
#features_data
#1.Temp represents the temperature of the region
#2.Unemployment represents the unemployment rate in the region
#3.CPI represents the consumer price index
#4.MarkDown represents the offer price or reduction in price
#5.FuelPrice represents the price of the fuel in that region
```

```python
features_data=pd.read_csv('/content/Features_data_set.csv')
features_data.head()
```

```python
features_data.isnull().sum()
```

```python
features_data[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']]=features_data[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']].fillna(value=0)
```

```python
features_data.isnull().sum()
```

```python
features_data[features_data['CPI'].isna()]
```

```python
features_data['CPI']=features_data['CPI'].fillna(value=features_data['CPI'].mean())
```

```python
features_data.isnull().sum()
```

```python
features_data['Unemployment']=features_data['Unemployment'].fillna(value=features_data['Unemployment'].mean())
```

```python
features_data.isnull().sum()
```

```python
#merging Sales_data and stores_data with respect to store
```

```python
sales_data_stores=pd.merge(sales_data,stores_data,on='Store')
```

```python
sales_data_stores.head()
```

```python
#merge features_data with the newly created  sales_stores dataframe
```

```python
sales_store_features=pd.merge(sales_data_stores,features_data,on=['Store','Date'])
```
